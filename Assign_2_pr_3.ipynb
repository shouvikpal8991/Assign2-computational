```python
import numpy as np
import matplotlib.pyplot as plt

def f1(y, yp, x):
    return yp

def f2(y, yp, x):
    return 2 * yp - y + x * np.exp(x) - x

def runge_kutta(f1, f2, h, a, b, y0, yp0):
    n = int((b - a) / h)
    x_values = [a]
    y_values = [y0]
    yp_values = [yp0]

    for i in range(n):
        x = x_values[-1]
        y = y_values[-1]
        yp = yp_values[-1]

        k1 = h * f1(y, yp, x)
        l1 = h * f2(y, yp, x)
        k2 = h * f1(y + k1/2, yp + l1/2, x + h/2)
        l2 = h * f2(y + k1/2, yp + l1/2, x + h/2)
        k3 = h * f1(y + k2/2, yp + l2/2, x + h/2)
        l3 = h * f2(y + k2/2, yp + l2/2, x + h/2)
        k4 = h * f1(y + k3, yp + l3, x + h)
        l4 = h * f2(y + k3, yp + l3, x + h)

        y_new = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        yp_new = yp + (l1 + 2*l2 + 2*l3 + l4) / 6
        x_new = x + h

        x_values.append(x_new)
        y_values.append(y_new)
        yp_values.append(yp_new)

    return x_values, y_values

# Initial conditions
a = 0
b = 1
h = 0.01
y0 = 0
yp0 = 0

x_values, y_values = runge_kutta(f1, f2, h, a, b, y0, yp0)

plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of the ODE using 4th-order Runge-Kutta Method')
plt.grid(True)
plt.show()
```
