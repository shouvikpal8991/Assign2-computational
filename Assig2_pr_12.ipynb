```python
import numpy as np
import matplotlib.pyplot as plt


def f(t, u):
    u1, u2, u3 = u
    du1dt = u1 + 2*u2 - 2*u3 + np.exp(-t)
    du2dt = u2 + u3 - 2*np.exp(-t)
    du3dt = u1 + 2*u2 + np.exp(-t)
    return np.array([du1dt, du2dt, du3dt])

def runge_kutta4(t0, u0, h, f):
    k1 = h * f(t0, u0)
    k2 = h * f(t0 + h/2, u0 + k1/2)
    k3 = h * f(t0 + h/2, u0 + k2/2)
    k4 = h * f(t0 + h, u0 + k3)
    return u0 + (k1 + 2*k2 + 2*k3 + k4) / 6

t0 = 0
u0 = np.array([3, -1, 1])

h = 0.01
num_steps = int(1 / h)

# Arrays to store results
t_values = np.linspace(t0, 1, num_steps+1)
u_values = np.zeros((num_steps+1, 3))

# Perform Runge-Kutta integration
u_values[0] = u0
for i in range(num_steps):
    u_values[i+1] = runge_kutta4(t_values[i], u_values[i], h, f)

# Plot the results
plt.plot(t_values, u_values[:, 0], label='u1(t)')
plt.plot(t_values, u_values[:, 1], label='u2(t)')
plt.plot(t_values, u_values[:, 2], label='u3(t)')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.title('Solution of the System of ODEs using Runge-Kutta 4th Order')
plt.legend()
plt.grid(True)
plt.show()
```
