```python
import numpy as np
import matplotlib.pyplot as plt


def euler_method(f,h,a,b,y0):
  n = int((b - a) / h)
  t = np.zeros(n+1)
  y = np.zeros(n+1)
  t[0] = a
  y[0] = y0

  for i in range(n): #it runs the loop from i=0 to i=n-1
      t[i+1]=t[i] + h
      y[i+1]=y[i] + h * f(y[i], t[i])
  return t,y


def f1(y, t):
    return (y / t) - (y / t) ** 2

t_values,y_numerical=euler_method(f1,0.1,1,2,1)


# Calculate y values analytically
y_analytic = [t / (1 + np.log(t)) for t in t_values]

plt.plot(t_values, y_numerical, label='Numerical (Euler Method)', color='blue')
plt.plot(t_values, y_analytic, label='Analytical', color='red')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Euler Method vs Analytical Solution')
plt.legend()
plt.grid(True)
plt.show()

# Convert y_analytic and y_numerical to numpy arrays for element-wise operations
y_analytic = np.array(y_analytic)
y_numerical = np.array(y_numerical)

# Calculate absolute error
absolute_errors = abs(y_analytic - y_numerical)

# Calculate relative error
relative_errors = absolute_errors / abs(y_analytic)

print("Absolute Errors:", absolute_errors)
print("Relative Errors:", relative_errors)
```
