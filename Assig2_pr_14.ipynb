```python
import numpy as np
import matplotlib.pyplot as plt

def f1(y,yp,t):
  return yp
def f2(y,yp,t):
  return ((2/t)*yp)-((2/t**2)*y)+(t*np.log(t))
def f3(t):
  return (((7*t)/4)+((t**3/2)*np.log(t))-(((3/4)*t**3)))

def euler_method(f1,f2,f3,a,b,h,y0,yp0):
  n = int((b - a) / h)
  t = np.zeros(n+1) #this means this array has n no of zeros
  y = np.zeros(n+1)
  yp = np.zeros(n+1)
  y_exact=np.zeros(n+1)
  t[0] = a
  y[0] = y0
  yp[0]= yp0
  y_exact[0]=y0

  for i in range(n): #it runs the loop from i=0 to i=n-1
      t[i+1]=t[i] + h
      yp[i+1]=yp[i] + h * f2(y[i],yp[i], t[i])
      y[i+1]=y[i] + h * f1(y[i],yp[i], t[i])
      y_exact[i+1]=f3(t[i+1])

  return t,y,y_exact

t,y,y_exact= euler_method(f1,f2,f3,1,2,0.001,1,0)

# Plot the results
plt.plot(t, y, label='Euler Method', color='blue')
plt.plot(t, y_exact, label='Exact Solution', color='red', linestyle='--')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution of the Initial Value Problem by euler method')
plt.legend()
plt.grid(True)
plt.show()
```
