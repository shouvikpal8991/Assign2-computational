```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect

def f1(x,y):
   return -9*y

def backward_euler(f1, a, y0, b, h):
  n=int((b-a)/h)
  x_values=np.zeros(n+1)
  x_values[0]=a
  #y_values=np.zeros(n+1)
  y_values= [y0]
  for i in range (1, n+1):
    x_values[i]=x_values[i-1]+h

    def g(y):
       return  y - y_values[i-1] - h*f1(x_values[i],y)
    y = bisect(g, -10, 10)
    y_values.append(y)

  return x_values,y_values

a=0
b=1
y0=np.exp(1)
h=0.01
X1, Y1 = backward_euler(f1, a, y0, b, h)

plt.plot(X1, Y1, color='g', label='Backward Euler Solution')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Solution to dy/dx=-9y, y(0)=e')
plt.legend()
plt.show()
#def true_sol(x):
  #return np.exp(1-9*x)
#y_true=true_sol(X1)
#plt.plot(X1,y_true)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


def f2(x,y):
   return -20*(y-x)**2+2*x

def backward_euler(f2, a, y0, b, h):
  n=int((b-a)/h)
  x_values=np.zeros(n+1)
  x_values[0]=a
  #y_values=np.zeros(n+1)
  y_values= [y0]
  for i in range (1, n+1):
    x_values[i]=x_values[i-1]+h

    def g(y):
       return  y - y_values[i-1] - h*f2(x_values[i],y)
    y = fsolve(g, y0)[0]
    y_values.append(y)

  return x_values,y_values

a=0
b=1
y0=(1/3)
h=0.01
X1, Y1 = backward_euler(f2, a, y0, b, h)

plt.plot(X1, Y1, color='r', label='Backward Euler Solution')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Solution to dy/dx=-20(y-x)^2+2x, y(0)=1/3')
plt.legend()
plt.show()
```
